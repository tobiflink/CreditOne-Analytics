```python
#Convert columns to category
credit_refined['Education'] = pd.Categorical(credit_refined.Education)
credit_refined['Sex'] = pd.Categorical(credit_refined.Sex)
credit_refined['Marriage'] = pd.Categorical(credit_refined.Marriage)
credit_refined['Default'] = pd.Categorical(credit_refined.Default)
```

```python
credit_refined.info()
```

```python
# importing one hot encoder from sklearn 
from sklearn.preprocessing import OneHotEncoder 
```

```python
#Gender

credit_refined['Male'] = credit_refined['Sex'].map( {1:1, 2:0} )
credit_refined[['Sex', 'Male']]
```

```python
#Education Column

#Replacing Education column with (1 = graduate school; 2 = university; 3 = high school; 0, 4, 5, 6 = others).

credit_refined["Education"].replace({ 1: "GradSchl", 2: "Uni", 3: "HighSchl", 4: "Others"}, inplace=True)

#Replacing Mariage column (1 = married; 2 = single; 3 = divorce; 0=others).

credit_refined["Marriage"].replace({ 0: "Others", 1: "Married", 2: "Single", 3: "Others"}, inplace = True)
```

```python
credit_refined.head()
```

```python
#One hot encoding Education
dummies1 = pd.get_dummies(credit_refined['Education'], prefix='')
```

```python
#One hot encoding Marriage
dummies2= pd.get_dummies(credit_refined['Marriage'], prefix='')
```

```python
print(dummies1)
```

```python
print(dummies2)
```

```python
#To concanate new features in data
credit_refined = pd.concat([credit_refined, dummies1, dummies2 ], axis=1)
```

```python
credit_refined.head()
```

```python
#Drop Sex, Education & Marriage columns

credit_refined.drop(["Sex", "Education", "Marriage"], axis=1, inplace=True)
```

```python
#Export data

credit_refined.to_csv(r'/Users/tobiadelokun/Documents/Ubiqum/Python/Credit_One_Credit_Analysis/Cleaned_Data.csv', index = False)
```
